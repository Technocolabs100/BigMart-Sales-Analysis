Hypothesis Generation

Item Type and Sales: Items of certain types might have higher sales than others. For example, items categorized as "Fruits and Vegetables" or "Meat" might have higher sales due to their essential nature in grocery shopping.

Item Fat Content and Sales: Items with different fat content categories (Low Fat, Regular, etc.) might have varying sales. Consumers might prefer healthier options, leading to higher sales for low-fat items.

Item Visibility and Sales: Items with higher visibility might have better sales, as they are more likely to catch the attention of shoppers.

Outlet Location and Sales: Outlets in different locations (Tier 1, Tier 2, Tier 3) might have varying sales due to differences in consumer demographics and preferences.

Outlet Size and Sales: The size of the outlet might influence sales, with larger outlets potentially accommodating more products and customers, leading to higher sales.

Outlet Type and Sales: Different outlet types (Supermarket Type1, Type2, Type3, Grocery Store) might have different sales patterns. Supermarkets with more variety might attract more customers and generate higher sales.

Outlet Establishment Year and Sales: Older outlets might have a stronger customer base and established reputation, potentially leading to higher sales.

Item Weight and Sales: There might be a relationship between the weight of an item and its sales. Heavier items might be more expensive and thus generate higher sales.

Outlet Establishment Year and Outlet Size: Older outlets might be larger and more established, leading to a correlation between the establishment year and the outlet size.

Outlet Location and Outlet Type: The type of outlet might be influenced by its location. For example, Tier 1 locations might have more Supermarket Type outlets.

Item MRP and Sales: Items with higher Maximum Retail Prices (MRP) might generate higher sales, as they could be perceived as higher quality or more luxurious.

Item Type and Item Visibility: Certain item types might be strategically placed in more visible areas of the store, leading to higher visibility.

Handle missing values

```python
# Impute missing values for 'Item_Weight' column using mean
mean_item_weight = data['Item_Weight'].mean()
data['Item_Weight'].fillna(mean_item_weight, inplace=True)

# Impute missing values for 'Outlet_Size' column using mode
mode_outlet_size = data['Outlet_Size'].mode()[0]
data['Outlet_Size'].fillna(mode_outlet_size, inplace=True)

# Print the first few rows of the modified DataFrame
print(data.head())
```

```python
data.isnull().sum()
```

Convert Categorical variables into features


```python
categorical_columns = ['Item_Fat_Content', 'Item_Type', 'Outlet_Identifier', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type']
data = pd.get_dummies(data, columns=categorical_columns, drop_first=True)
```

```python
data.head(10)
```

```python
# Assuming the item identifiers follow the pattern "AAA123", where AAA is the category code
data['Item_Category'] = data['Item_Identifier'].str[:3]
```

```python
print(data.columns)
```

```python
data.head(10)
```

```python
categorical_columns = ['Item_Category']
data = pd.get_dummies(data, columns=categorical_columns, drop_first=True)
```

```python
data.head(20)
```

```python
X=data.drop(['Item_Outlet_Sales','Item_Identifier'],axis=1)
```

```python
Y=data['Item_Outlet_Sales']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
standard=scaler.fit(X)
```

```python
X=scaler.transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,mean_squared_error
pred=model.predict(X_test)
accuracy=metrics.r2_score(Y_test,pred)*100
accuracy
```
